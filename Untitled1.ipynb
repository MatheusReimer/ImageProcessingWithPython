```python
import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
from __future__ import division


url = "https://images.unsplash.com/photo-1554273186-c3282aa9285f?ixlib=rb-1.2.1&ixid=MnwxMjA3fDB8MHxleHBsb3JlLWZlZWR8NDd8fHxlbnwwfHx8fA%3D%3D&auto=format&fit=crop&w=500&q=60"
url2= "https://images.unsplash.com/photo-1533035353720-f1c6a75cd8ab?ixid=MnwxMjA3fDB8MHxzZWFyY2h8MXx8Z3JheSUyMGJhY2tncm91bmR8ZW58MHx8MHx8&ixlib=rb-1.2.1&auto=format&fit=crop&w=500&q=60"

###################################################################################################################################################################################################
def preLoop():
  image = io.imread(url) 

  ### ORIGINAL IMAGE

  ##cv2_imshow(image) 
  image_data = np.asarray(image)
  image_data = image_data.astype(np.float64)
  return image_data

###################################################################################################################################################################################################
def grayscaleAritmetica():
  ##Prints original first
  
  image_data = preLoop()
  cv2_imshow(image_data) 
  ## Iterate through every pixel and every R/G/B, then uses the average of it to display it


  for i in range(len(image_data)):
      for j in range(len(image_data[0])):
          #######print(image_data[i][j])  # this row prints an array of RGB color for each pixel in the image
          avg = 0
          avg += image_data[i, j, 0]
          avg += image_data[i, j, 1]
          avg += image_data[i, j, 2]
          avg /=3
          image_data[i, j] = avg
  cv2_imshow(image_data) 


#####grayscaleAritmetica()



###########################################################################################################################################################################################
def grayscalePonderada():
    image_data = preLoop()

    for i in range(len(image_data)):
      for j in range(len(image_data[0])):
          #######print(image_data[i][j])  # this row prints an array of RGB color for each pixel in the image
          avg = 0
          avg += image_data[i, j, 0] * 0.1
          avg += image_data[i, j, 1] * 0.2
          avg += image_data[i, j, 2] * 0.7
        
          image_data[i, j] = avg
    cv2_imshow(image_data) 


###grayscalePonderada()




###########################################################################################################################################################################################
def expressions(operation):
  image_data = preLoop()
  image_to_compare = io.imread(url2) 

  for i in range(len(image_data)):
      for j in range(len(image_data[0])):
 
          if operation =="+":
            image_data[i, j] += image_to_compare[i, j]
          elif operation == "*":
            image_data[i, j] *= image_to_compare[i, j]
          elif operation == "/":
            image_data[i, j] /= image_to_compare[i, j]
          elif operation == "-":
            image_data[i, j] -= image_to_compare[i, j]
          else:
            image_data[i, j] =  image_data[i, j]
  cv2_imshow(image_data) 




###expressions("+")
###expressions("*")
###expressions("/")
###expressions("-")

###################################################################################################################################################################################################


def bgrToZero(color):
  image_data = preLoop()
  for i in range(len(image_data)):
      for j in range(len(image_data[0])):
        if color=="B":
          image_data[i, j, 1] = 0
          image_data[i, j, 2] = 0
        elif color=="G":
          image_data[i, j, 0] = 0
          image_data[i, j, 2] = 0
        elif color=="R":
          image_data[i, j, 0] = 0
          image_data[i, j, 1] = 0
  cv2_imshow(image_data) 


##bgrToZero("B")
##bgrToZero("G")
##bgrToZero("R")


########################################################################################################################################################


def threshold(limit):
  image_data = preLoop()
  for i in range(len(image_data)):
      for j in range(len(image_data[0])):
        
        avg = image_data[i, j, 0] + image_data[i, j, 1] + image_data[i, j, 2]
        avg /= 3
        if avg>=limit:
          image_data[i, j, 0],image_data[i, j, 1], image_data[i, j, 2] = 255, 255, 255
        else:
          image_data[i, j, 0],image_data[i, j, 1], image_data[i, j, 2] = 0, 0, 0
  cv2_imshow(image_data)






#########################################################################################################################################################


def convolution():
    image_data = preLoop()
    numrows = len(image_data) 
    numcols = len(image_data) 
    print(numrows, numcols)
    for i in range(len(image_data)):
      for j in range(len(image_data[0])):
        ### Dealing with Overflow in X to the right
        if i+1>=numrows:
          (i) = i-1
        ### Dealing with Overflow in X to the left          
        if i-1<0:
          i = i+1
        ### Dealing with Overflow in Y to the top
        if j+1>=numcols:
          j = j-1
        ### Dealing with Overflow in Y to the bottom
        if j-1<0:
          j = j+1
        
        midle = image_data[i,j] 
        a11 = image_data[i-1,j-1] 
        a12 = image_data[i,j-1]
        a13 = image_data[i+1,j-1]
        a21 = image_data[i-1,j]
        a23 = image_data[i+1,j]
        a31 = image_data[i-1,j+1]
        a32 = image_data[i,j+1]
        a33 = image_data[i+1,j+1]  
        image_data[i,j] = (a11+a12+a13+a21+a23+a31+a32+a33+midle)/9

    cv2_imshow(image_data)
    


##############################################################################################################################################################
from copy import copy, deepcopy





def erosion():
  urlToErosion = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcR66J9pkm63Nz_Zsx15uvD9_KBY_nWfaMhk0g&usqp=CAU"
  image = io.imread(urlToErosion) 

  ### ORIGINAL IMAGE

  ##cv2_imshow(image) 
  image_data = np.asarray(image)
  image_data = image_data.astype(np.float64)
  numrows = len(image_data) 
  numcols = len(image_data)
  cv2_imshow(image_data)
  ##COPYING TO ANOTHER 2D ARRAY
  erodedMatrix = deepcopy(image_data)
  

  for i in range(1,(len(image_data))-1):
      for j in range(1,(len(image_data[0]))-1):
        a12 = image_data[i,j+1]
        a21 = image_data[i-1,j]
        a23 = image_data[i+1,j]
        a32 = image_data[i,j-1]
     
   
        if a12 > 1 and a21 > 1 and a23 > 1 and a32 > 130 :
          erodedMatrix[i,j] = 255
        else:
          erodedMatrix[i,j] = 0
    




        

  cv2_imshow(erodedMatrix)





####################################################################################################################################################
def dilation():
  urlToDilation = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTHWkZoWEHpDVzyA-l4xlgiIjvH2vjnTb12sA&usqp=CAU"
  image = io.imread(urlToDilation) 

  ### ORIGINAL IMAGE

  ##cv2_imshow(image) 
  image_data = np.asarray(image)
  image_data = image_data.astype(np.float64)
  numrows = len(image_data) 
  numcols = len(image_data)
  cv2_imshow(image_data)
  ##COPYING TO ANOTHER 2D ARRAY
  dilatedMatrix = deepcopy(image_data)
  

  for i in range(1,(len(image_data))-1):
      for j in range(1,(len(image_data[0]))-1):

     
        
        if image_data[i,j] >1 :
          dilatedMatrix[i,j+1] = 255
          dilatedMatrix[i-1,j] = 255
          dilatedMatrix[i+1,j] = 255
          dilatedMatrix[i,j-1] = 255
        else:
          dilatedMatrix[i,j+1] = 0
          dilatedMatrix[i-1,j] = 0
          dilatedMatrix[i+1,j] = 0
          dilatedMatrix[i,j-1] = 0




        

  cv2_imshow(dilatedMatrix)








##############################################################################################################################################


def opening():
  urlToDilation = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTHWkZoWEHpDVzyA-l4xlgiIjvH2vjnTb12sA&usqp=CAU"
  image = io.imread(urlToDilation) 

  ### ORIGINAL IMAGE

  ##cv2_imshow(image) 
  image_data = np.asarray(image)
  image_data = image_data.astype(np.float64)
  numrows = len(image_data) 
  numcols = len(image_data)
  cv2_imshow(image_data)
  erodedMatrix = deepcopy(image_data)
  ##COPYING TO ANOTHER 2D ARRAY
  
  
  ###EROSION
  for i in range(1,(len(image_data))-1):
      for j in range(1,(len(image_data[0]))-1):

        a12 = image_data[i,j+1]
        a21 = image_data[i-1,j]
        a23 = image_data[i+1,j]
        a32 = image_data[i,j-1]
     
   
        if a12 > 1 and a21 > 1 and a23 > 1 and a32 > 130 :
          erodedMatrix[i,j] = 255
        else:
          erodedMatrix[i,j] = 0
  ##DILATION
  dilatedMatrix = deepcopy(erodedMatrix)
  for i in range(1,(len(erodedMatrix))-1):
      for j in range(1,(len(erodedMatrix[0]))-1):

     
        
        if image_data[i,j] >1 :
          dilatedMatrix[i,j+1] = 255
          dilatedMatrix[i-1,j] = 255
          dilatedMatrix[i+1,j] = 255
          dilatedMatrix[i,j-1] = 255
        else:
          dilatedMatrix[i,j+1] = 0
          dilatedMatrix[i-1,j] = 0
          dilatedMatrix[i+1,j] = 0
          dilatedMatrix[i,j-1] = 0



        

  cv2_imshow(dilatedMatrix)





dilation()











```
