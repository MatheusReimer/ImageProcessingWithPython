```python
import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
from __future__ import division


url = "https://images.unsplash.com/photo-1554273186-c3282aa9285f?ixlib=rb-1.2.1&ixid=MnwxMjA3fDB8MHxleHBsb3JlLWZlZWR8NDd8fHxlbnwwfHx8fA%3D%3D&auto=format&fit=crop&w=500&q=60"
url2= "https://images.unsplash.com/photo-1585314062340-f1a5a7c9328d?ixid=MnwxMjA3fDB8MHxzZWFyY2h8N3x8Z3JheXxlbnwwfHwwfHw%3D&ixlib=rb-1.2.1&auto=format&fit=crop&w=500&q=60"

###################################################################################################################################################################################################
def preLoop():
  image = io.imread(url) 

  ### ORIGINAL IMAGE

  ##cv2_imshow(image) 
  image_data = np.asarray(image)
  image_data = image_data.astype(np.float64)
  return image_data

###################################################################################################################################################################################################
def grayscaleAritmetica():
  ##Prints original first
  
  image_data = preLoop()
  cv2_imshow(image_data) 
  ## Iterate through every pixel and every R/G/B, then uses the average of it to display it


  for i in range(len(image_data)):
      for j in range(len(image_data[0])):
          #######print(image_data[i][j])  # this row prints an array of RGB color for each pixel in the image
          avg = 0
          avg += image_data[i, j, 0]
          avg += image_data[i, j, 1]
          avg += image_data[i, j, 2]
          avg /=3
          image_data[i, j] = avg
  cv2_imshow(image_data) 


grayscaleAritmetica()



###########################################################################################################################################################################################
def grayscalePonderada():
    image_data = preLoop()

    for i in range(len(image_data)):
      for j in range(len(image_data[0])):
          #######print(image_data[i][j])  # this row prints an array of RGB color for each pixel in the image
          avg = 0
          avg += image_data[i, j, 0] * 0.1
          avg += image_data[i, j, 1] * 0.2
          avg += image_data[i, j, 2] * 0.7
        
          image_data[i, j] = avg
    cv2_imshow(image_data) 


grayscalePonderada()




###########################################################################################################################################################################################
def expressions(operation):
  image_data = preLoop()
  image_to_compare = io.imread(url2) 

  for i in range(len(image_data)):
      for j in range(len(image_data[0])):
 
          if operation =="+":
            image_data[i, j] += image_to_compare[i, j]
          elif operation == "*":
            image_data[i, j] *= image_to_compare[i, j]
          elif operation == "/":
            image_data[i, j] /= image_to_compare[i, j]
          elif operation == "-":
            image_data[i, j] -= image_to_compare[i, j]
          else:
            image_data[i, j] =  image_data[i, j]
  cv2_imshow(image_data) 




expressions("+")
expressions("*")
expressions("-")
expressions("/")
###################################################################################################################################################################################################


def bgrToZero(color):
  image_data = preLoop()
  for i in range(len(image_data)):
      for j in range(len(image_data[0])):
        if color=="B":
          image_data[i, j, 1] = 0
          image_data[i, j, 2] = 0
        elif color=="G":
          image_data[i, j, 0] = 0
          image_data[i, j, 2] = 0
        elif color=="R":
          image_data[i, j, 0] = 0
          image_data[i, j, 1] = 0
  cv2_imshow(image_data) 


bgrToZero("B")
bgrToZero("G")
bgrToZero("R")
```
